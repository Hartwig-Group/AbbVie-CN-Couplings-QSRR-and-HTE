```python
import pandas as pd
from itertools import product
import numpy as np
```

```python
dfnuc_triaged = pd.read_csv("selected_nucleophiles_v4-purchased.csv",index_col=0,header=0)
dfel_triaged = pd.read_csv("selected_electrophiles_v4-purchased.csv",index_col=0,header=0)

dfnuc_triaged.drop(dfnuc_triaged.index[dfnuc_triaged['RMM Availability'] != 'Yes'],inplace=True)
dfel_triaged.drop(dfel_triaged.index[dfel_triaged['RMM Availability'] != 'Yes'],inplace=True)

nuc_smi = dfnuc_triaged['SMILES String']
el_smi = dfel_triaged['SMILES String']
nuc_lbl = dfnuc_triaged['CAS Number']
el_lbl = dfel_triaged['CAS Number']


# couplings_all = [f for f in product(nucs,elecs)] # CHANGED - triaged electrophiles
couplings_all = [f for f in product(list(zip(nuc_lbl,nuc_smi)),list(zip(el_lbl,el_smi)))]

print(couplings_all)

np.random.shuffle(couplings_all)
# print(couplings_all)
print(len(couplings_all))

## Getting 220 combinations
nucs = [f[0] for f in couplings_all[:220]]
elcs = [f[1] for f in couplings_all[:220]]
# # Number of unique reactants contained in the set of experiments above
# print(pd.Series(nucs).nunique())
# print(pd.Series(elcs).nunique())
# print(elcs)
# ellbls,elsmi = zip(*el_final_selection)

import rdkit.Chem.Draw as rddraw
import rdkit.Chem.rdmolfiles as rdm

# drawing = rddraw.MolsToGridImage(mols = [rdm.MolFromSmiles(f) for f in el_smi],molsPerRow=4,subImgSize=(400,400),useSVG=True,returnPNG=False,legends=ellbls,maxMols=100)

# with open(f"triaged_electrophiles.svg",'w') as k:
#     k.write(drawing.data)
from rdkit.Chem.Draw import MolDraw2DSVG

test = []
labels = []
for n,e in zip(nucs,elcs):
    test.append(n[1])
    test.append(e[1])
    labels.append(n[0])
    labels.append(e[0])
print(len(test),len(labels))
# print(test[:10])
# # break
drawing = rddraw._MolsToGridSVG(mols=[rdm.MolFromSmiles(f) for f in test],molsPerRow=6,subImgSize=(400,400),legends=labels)
print(type(drawing))
with open(f"coupling_strucs_all_v4.svg",'w') as k:
    k.write(drawing)
```

```python
ncas=[]
ecas=[]
nsmi=[]
esmi=[]

for n,e in zip(nucs,elcs):
    nsmi.append(n[1])
    esmi.append(e[1])
    ncas.append(n[0])
    ecas.append(e[0])
    
# print(ecas)
# print(couplings_all)
```

```python
plate1df = pd.DataFrame()
idx = [''.join(f) for f in product(('a','b','c','d','e','f','g','h'),[str(f+1) for f in range(12)])]
plate1df.index = idx
plate1df['Nucleophile CAS'] = ncas[:96]
plate1df['Nucleophile SMILES'] = nsmi[:96]
plate1df['Electrophile CAS'] = ecas[:96]
plate1df['Electrophile SMILES'] = esmi[:96]

print(plate1df.head)

plate1df.to_csv("plate1_design.csv",header=True)
```
